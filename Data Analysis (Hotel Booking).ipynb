```python
df['Month'] = pd.to_datetime(df['reservation_status_date']).dt.strftime('%B')
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
plt.figure(figsize=(16, 8))
ax1 = sns.countplot(x='Month', hue='is_canceled', data=df, order=month_order, palette='Set1')
plt.xlabel('Month', fontsize=18)
plt.ylabel('Number of Reservations', fontsize=18)
plt.legend(['Not Cancelled', 'Cancelled'], fontsize=14)
plt.title('Month-wise Cancellation', fontsize=22, fontweight='bold')

plt.show()
```

```python
# Assuming 'df' is your DataFrame and 'Month' is your column with month names
# Define the correct month order
plt.style.use('seaborn-v0_8-whitegrid')
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']

# Convert the 'Month' column to a categorical type with the specified order
df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)

# Group by Month and sum the ADR (Average Daily Rate) for canceled bookings
Avg_price_month = df[df['is_canceled'] == 1].groupby('Month')['adr'].sum().reset_index()

# Plotting using Seaborn for better aesthetics
plt.figure(figsize=(15,8))
plt.title('Total ADR by Month for Canceled Bookings', fontsize=30)
sns.barplot(x='Month', y='adr', data=Avg_price_month, color='skyblue')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Total ADR', fontsize=14)
plt.xticks(rotation=45)

plt.show()
```

```python
cancelled_df=df[df['is_canceled']==1]
not_cancelled_df=df[df['is_canceled']==0]
```

```python
cancelled_df['market_segment'].value_counts(normalize=True)
```

```python

# Assuming reservation_status_date is a string in the format 'YYYY-MM-DD'
cancelled_df['reservation_status_date'] = pd.to_datetime(cancelled_df['reservation_status_date'])
not_cancelled_df['reservation_status_date'] = pd.to_datetime(not_cancelled_df['reservation_status_date'])

# Define the date range from January 2016 to September 2017
start_date = '2016-01-01'
end_date = '2017-09-30'

# Filter the data based on the date range
cancelled_df_filtered = cancelled_df[(cancelled_df['reservation_status_date'] >= start_date) & 
                                     (cancelled_df['reservation_status_date'] <= end_date)]

not_cancelled_df_filtered = not_cancelled_df[(not_cancelled_df['reservation_status_date'] >= start_date) & 
                                             (not_cancelled_df['reservation_status_date'] <= end_date)]

# Calculate the average ADR for canceled reservations
cancelled_df_adr = cancelled_df_filtered.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Calculate the average ADR for not canceled reservations
not_cancelled_df_adr = not_cancelled_df_filtered.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Plot the ADR for both canceled and not canceled reservations
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate (ADR) by Reservation Status (2016 to September 2017)', fontsize=25)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled', color='green', linewidth=2)
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled', color='red')
plt.xlabel('Reservation Status Date', fontsize=18)
plt.ylabel('ADR', fontsize=18)
plt.legend(title='Reservation Status', fontsize=15, title_fontsize='20')
plt.grid(True)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()

```
